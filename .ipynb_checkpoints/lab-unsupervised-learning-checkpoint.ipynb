# Lab | Unsupervised Learning

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from tqdm import tqdm
from sklearn.metrics import silhouette_score
```

```python
data = pd.read_csv('combined_tracks.csv')
data
```

```python
data.columns.to_list()
```

```python
# Selecting the relevant audio features for clustering
features = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 
            'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
```

```python
X = data[features]
```

```python
# Splitting the dataset into training and testing sets (80% train, 20% test)
X_train, X_test, train_indices, test_indices = train_test_split(X, data.index, test_size=0.2, random_state=1989)
```

```python
# Fit the scaler on the training data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# Transform the testing data using the fitted scaler
X_test_scaled = scaler.transform(X_test)
```

# Elbow Method

```python
inertia = []
cluster_range = range(1, 100)  # Checking for up to 50 clusters

for cluster_num in tqdm(cluster_range):
    kmeans = KMeans(n_clusters=cluster_num, n_init=10, random_state=1989)
    kmeans.fit(X_test_scaled)
    inertia.append(kmeans.inertia_)

# Plotting the elbow curve
plt.figure(figsize=(12, 5))
plt.plot(cluster_range, inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.grid(True)
plt.show()
```

# Silhouette method

```python
silhouette_scores = []

cluster_range_silhouette = range(2, 100)  # silhouette score requires at least 2 clusters

# Calculate silhouette scores for different numbers of clusters
for cluster_num in tqdm(cluster_range_silhouette):
    kmeans = KMeans(n_clusters=cluster_num, n_init=10, random_state=1989)
    cluster_labels = kmeans.fit_predict(X_test_scaled)
    silhouette_avg = silhouette_score(X_test_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plotting the silhouette scores
plt.figure(figsize=(12, 5))
plt.plot(cluster_range_silhouette, silhouette_scores, marker='o', linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('Average Silhouette Score')
plt.title('Silhouette Analysis')
plt.grid(True)
plt.show()
```

```python
kmeans_41 = KMeans(n_clusters=41, random_state=1989, n_init=10)
X_train_clusters = kmeans_41.fit_predict(X_train_scaled)
X_test_clusters = kmeans_41.predict(X_test_scaled)

cluster_distribution = np.bincount(X_train_clusters)

cluster_distribution
```

```python
silhouette_train = silhouette_score(X_train, X_train_clusters)
silhouette_test = silhouette_score(X_test, X_test_clusters)
```

```python
silhouette_train, silhouette_test
```

Close Scores: The silhouette scores for both the training and test sets are very close. This suggests that the quality and structure of the clusters in the training data are consistent with the test data. It indicates that the clustering generalizes well to unseen data.

Negative Scores: Silhouette scores can range from -1 to 1. A high value indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters. A negative value indicates that samples might have been assigned to the wrong clusters. In our case, the negative values suggest there might be some overlap between clusters. However, since the values are close to zero, the overlap might not be very significant.

Practical Implication: Given that the scores are consistent between the training and test sets, we can be reasonably confident in using the clusters for recommendation. However, the negative scores suggest that there might be room for improving the clustering.
