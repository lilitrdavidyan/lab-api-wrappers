```python
def print_json_structure(data, indent=0):
    if isinstance(data, dict):
        for key, value in data.items():
            print(" " * indent + f"- {key}")
            if isinstance(value, (dict, list)):
                print_json_structure(value, indent + 2)
    elif isinstance(data, list):
        for item in data:
            if isinstance(item, (dict, list)):
                print_json_structure(item, indent + 2)
```

## Fetching tracks from each playlist

```python
def get_tracks_from_playlist(playlist_id):
    offset = 0
    tracks = []
    while True:
        response = sp.playlist_tracks(playlist_id, offset=offset)
        tracks.extend(response['items'])
        if response['next'] is None:
            break
        offset += 100
        time.sleep(0.5)
    return tracks
```

```python
def get_all_tracks(playlist_ids):
    all_tracks = []
    for playlist_id in playlist_ids:
        tracks = get_tracks_from_playlist(playlist_id)
        all_tracks.extend(tracks)
#         all_tracks.extend([extract_track_info(track) for track in tracks])
    return all_tracks
```

## Retrieve Audio Features

```python
def get_audio_features_for_tracks(track_ids):
    audio_features_list = []
    batch_size = 100  

    for i in range(0, len(track_ids), batch_size):
        batch = track_ids[i:i + batch_size]
        audio_features_batch = sp.audio_features(batch)
        audio_features_list.extend(audio_features_batch)
        time.sleep(0.5)  

    return audio_features_list
```

```python
# Extracting all the song IDs from the simplified_tracks list
track_ids = [track['id'] for track in simplified_tracks]

# Fetching the audio features for all tracks
all_audio_features = get_audio_features_for_tracks(track_ids)
```

## Getting more songs

```python
def fetch_songs_by_year(year, limit):
    offset = 0
    tracks = []
    
    while len(tracks) < limit:
        try:
            results = sp.search(q=f'year:{year}', type='track', limit=50, offset=offset)
            
            # If no more tracks are returned, break the loop for this year
            if not results['tracks']['items']:
                break

            tracks.extend(results['tracks']['items'])
            offset += 50  # Only increment offset if the request was successful
            
        except spotipy.SpotifyException as e:
            if e.http_status == 429:
                # If we hit the rate limit, wait the specified duration and then continue
                retry_after = int(e.headers.get("Retry-After", 10))
                print(f"Rate limit reached. Retrying after {retry_after} seconds...")
                time.sleep(retry_after)
                continue  # Retry the same request by not modifying the offset
            else:
                print(f"Could not fetch {year}-{offset} due to an error: {e}")
                offset += 50  # Increment offset to skip this batch and try the next one
    return tracks
```

```python
all_tracks = []
years = list(range(1950, 2010))

for year in years:
    song_limit = 1000
    print(f"Fetching up to {song_limit} songs for {year}...")
    all_tracks.extend(fetch_songs_by_year(year, song_limit))
```

```python
def simplify_tracks(all_tracks):
    simplified_tracks = []
    artist_song_count = {}  

    for track_info in all_tracks:
        try:
            song_title = track_info['name']
            song_id = track_info['id']
            artists = ", ".join([artist['name'] for artist in track_info['artists']])
            artists_ids = ", ".join([artist['id'] for artist in track_info['artists']])
            album_name = track_info['album']['name']
            album_id = track_info['album']['id']
            year = track_info['album']['release_date'].split('-')[0]
            
            # Update song counts and collect artist IDs for each artist
            for artist in track_info['artists']:
                artist_name = artist['name']
                artist_id = artist['id']
                
                # Increment song count for the artist and store the artist's ID
                if artist_name in artist_song_count:
                    artist_song_count[artist_name]['count'] += 1
                else:
                    artist_song_count[artist_name] = {'count': 1, 'id': artist_id}

            simplified_tracks.append({
                "title": song_title,
                "artists": artists,
                "artists_ids": artists_ids,
                "id": song_id,
                "album": album_name,
                "album_id": album_id,
                "year": year
            })
        except Exception as e:
            print(f"Error: {e}")
            print(track_info)
    return simplified_tracks, artist_song_count
```
